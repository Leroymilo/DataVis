````python


# Data preparation
japan_temperature = (city_temperature
    .query('Country == "Japan" and Year == 2010')                 # filter()
    .assign(Date=lambda x: pd.to_datetime(                        # mutate() with as.Date()
        x[['Year', 'Month', 'Day']].astype(str).agg('-'.join, axis=1)
    ))
    .loc[:, ['City', 'Date', 'AvgTemperature']]                  # select()
    .reset_index(drop=True)
)

print(japan_temperature)
````

````python
import matplotlib.dates as mdates

plt.figure(figsize=(6, 4))
sns.lineplot(data=japan_temperature,
            x='Date',
            y=japan_temperature['AvgTemperature'].rolling(30).mean(),  # Smoothed temperature
            hue='City')

# Format x-axis
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%b %Y'))

# Add light grey grid for y-axis
plt.grid(color='lightgrey', linestyle='-')
plt.tight_layout()
plt.show()
````

````python
# Convert dates to months and create pivot table
month_order = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
pivot_data = japan_temperature.pivot_table(
    values='AvgTemperature',
    index='City',
    columns=japan_temperature['Date'].dt.strftime('%b'),
    aggfunc='mean'
)[month_order]

# Create heatmap
plt.figure(figsize=(7, 3))
sns.heatmap(pivot_data,
            cmap='RdBu_r',
            center=50,
            annot=True,  # Show values in cells
            fmt='.1f',   # Round to 1 decimal place
            cbar_kws={'label': 'Temperature (F)'})

# Remove axis labels
plt.xlabel('')  # Remove x-axis label
plt.ylabel('')  # Remove y-axis label
plt.title('Average Temperature by City and Month in 2010')
plt.tight_layout()
plt.show()
````

````python
data = life_expectancy[life_expectancy['Year'] == 2010]
data = data[['Country', 'Population_mln']].dropna()

# Making a new column for the ratio of population growth to the world's median
data['Population_Ratio'] = data['Population_mln'] / data['Population_mln'].median()

# Reorder countries by Population_Ratio in descending order
data = data.sort_values(by='Population_Ratio', ascending=False)
data['Country'] = pd.Categorical(data['Country'], categories=data['Country'], ordered=True)

data
````

````python
print(BAD)

# Set plot size
plt.figure(figsize=(8, 6), dpi=300)

# Create the base scatter plot
sns.scatterplot(x='Country', y='Population_Ratio', data=data)

# Highlight specific countries
highlight_countries = data[data['Country'].isin(['Japan', 'China', 'Indonesia'])]

# Add text labels for specific countries
for i, row in highlight_countries.iterrows():
    plt.text(row['Country'], row['Population_Ratio']+5, row['Country'],
             horizontalalignment='left', verticalalignment='bottom', size=12)

# Add segments for specific countries
for i, row in highlight_countries.iterrows():
    plt.plot([row['Country'], row['Country']], [row['Population_Ratio'], row['Population_Ratio'] + 5],
             color='black')

# Customizing the plot
plt.xlabel('Countries, from most to least growth of population')
plt.ylabel('Population Growth / median')

# Remove x-axis text and ticks
plt.xticks([])
plt.tick_params(axis='x', which='both', bottom=False, top=False)

# Remove grid lines along the x-axis
plt.grid(False, axis='x')

plt.show()
````

````python
import calendar

# Filter the data for 2010 and specific cities
data = city_temperature[(city_temperature['Year'] == 2010) & (city_temperature['City'].isin(['Sydney', 'Tokyo', 'London', 'Paris', 'Beijing', 'Jakarta']))]

# Drop unnecessary columns
data = data.drop(['State', 'Country', 'Region'], axis=1)

# Create a new column called "Date"
data['Date'] = pd.to_datetime(data[['Year', 'Month', 'Day']].apply(lambda row: f"{row['Year']}-{row['Month']}-{row['Day']}", axis=1))

# Create a dataset with average temperature by month and city
heatmap_data = data.groupby(['Month', 'City'])['AvgTemperature'].mean().reset_index()

# Specify the order of the month names
heatmap_data['MonthName'] = heatmap_data['Month'].apply(lambda x: calendar.month_name[x])

# Map month names to month abbreviations
month_abbreviations = {name: abbreviation for name, abbreviation in zip(calendar.month_name[1:], calendar.month_abbr[1:])}
heatmap_data['MonthAbbreviation'] = heatmap_data['MonthName'].map(month_abbreviations)

# Sort the cities in the desired order
city_order = ['Sydney', 'London', 'Paris', 'Tokyo', 'Beijing', 'Jakarta']
heatmap_data['City'] = pd.Categorical(heatmap_data['City'], categories=city_order, ordered=True)
````
